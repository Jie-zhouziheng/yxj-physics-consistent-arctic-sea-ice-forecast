```python
import numpy as np
import xarray as xr
from pathlib import Path

# 路径
PROJECT_ROOT = Path.cwd().parent
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

# ======================
# 构造 toy 数据的维度
# ======================
n_years = 10
n_months = n_years * 12     # 120 个月
lat = np.linspace(60, 90, 20)   # 20 个纬度点
lon = np.linspace(-180, 180, 40)  # 40 个经度点
time = np.arange(
    np.datetime64("1980-01", "M"),
    np.datetime64("1990-01", "M"),
    dtype="datetime64[M]"
)
print("time length:", len(time), "first:", time[0], "last:", time[-1])

```

```python
# ======================
# 构造一个“像海冰”的场
# ======================
T, Y, X = len(time), len(lat), len(lon)

# 季节周期（冬高夏低）
season = 0.5 + 0.4 * np.cos(2 * np.pi * (np.arange(T) % 12) / 12)

# 空间结构（高纬冰多）
lat_factor = (lat - lat.min()) / (lat.max() - lat.min())

sic = np.zeros((T, Y, X), dtype="float32")

for t in range(T):
    sic[t] = (
        season[t]
        * lat_factor[:, None]
        + 0.05 * np.random.randn(Y, X)
    )

# 截断到 [0, 1]
sic = np.clip(sic, 0, 1)
```

```python
ds = xr.Dataset(
    {
        "sic": (("time", "lat", "lon"), sic)
    },
    coords={
        "time": time,
        "lat": lat,
        "lon": lon,
    }
)

out_path = PROCESSED_DIR / "toy_10yr_zscore.nc"
ds.to_netcdf(out_path)

print("Toy dataset saved to:", out_path)
print(ds)
```
